```python
# Live Streaming ----- > Auto Loader

## How to read Table as DataFrame and show(), display()
dfpdetails = spark.sql('select * from learning_catalog.learning_schema.dfpatient')
# dfpdetails.show()
# dfpdetails.display()
display(dfpdetails)
```

```python
%sql

-- CREATE SCHEMA learning_catalog.learning_Bronze;
-- CREATE SCHEMA learning_catalog.learning_Silver;
-- CREATE SCHEMA learning_catalog.learning_Gold;
```

```python
from pyspark.sql import functions as F
dfpatientClean = dfpdetails.selectExpr("CAST(patient_id AS INT) as patient_id", 
                                      "UPPER(TRIM(first_name)) as first_name", 
                                      "UPPER(TRIM(last_name)) as last_name", 
                                      "UPPER(TRIM(gender)) as gender", 
                                      "CAST(date_of_birth AS DATE) as date_of_birth", 
                                      "UPPER(TRIM(state_code)) as state_code", 
                                      "UPPER(TRIM(city)) as city", 
                                      "CAST(phone AS LONG) as phone",
                                      "CURRENT_DATE() as LoadDate"
                                      )


#dfpatientClean.display()

#'patient_id', 'first_name', 'last_name', 'gender', 'date_of_birth', 'state_code', 'city', 'phone'

dfpatientClean.write.mode("overwrite").saveAsTable("learning_catalog.learning_Bronze.dfpatients")
```
